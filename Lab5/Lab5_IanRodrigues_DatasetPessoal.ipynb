```python
#Projeto de Visualização 3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("dataset_ian.csv")

qualidade_sono = data["Qualidade do sono (0-10)"]
energia_dia_seguinte = data["Horas de sono"]

plt.figure(figsize=(10, 6))
plt.scatter(qualidade_sono, energia_dia_seguinte, color='lightskyblue', s=100, alpha=0.8, edgecolors='black', linewidths=0.5)
plt.title("Relação entre Qualidade do sono e Energia no dia seguinte")
plt.xlabel("Qualidade do sono")
plt.ylabel("Horas de sono")
plt.grid(False)

z = np.polyfit(qualidade_sono, energia_dia_seguinte, 1)
p = np.poly1d(z)

plt.show()
```

```python
#Projeto de Visualização 3
#Adicionando linha de tendência
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("dataset_ian.csv")

qualidade_sono = data["Qualidade do sono (0-10)"]
energia_dia_seguinte = data["Horas de sono"]

plt.figure(figsize=(10, 6))
plt.scatter(qualidade_sono, energia_dia_seguinte, color='lightskyblue', s=100, alpha=0.8, edgecolors='black', linewidths=0.5)
plt.title("Relação entre Qualidade do sono e Horas dormidas")
plt.xlabel("Qualidade do sono")
plt.ylabel("Horas dormidas")
plt.grid(False)

z = np.polyfit(qualidade_sono, energia_dia_seguinte, 1)
p = np.poly1d(z)

plt.plot(qualidade_sono, p(qualidade_sono), color="dodgerblue", linewidth=2, alpha=0.8)
plt.show()
```
